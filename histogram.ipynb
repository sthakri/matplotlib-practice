# Histograms in Matplotlib

This notebook demonstrates how to create histograms for visualizing data distribution:
- Basic histogram with custom bins
- Histogram with average line
- Styling and customization

## 1. Basic Histogram - Student Marks Distribution

```python
import matplotlib.pyplot as plt


data = [
    74, 88, 76, 65, 57, 57, 95, 67, 95, 55,
    85, 25, 52, 80, 62, 94, 89, 33, 30, 45,
    42, 25, 35, 75, 54, 81, 98, 33, 23, 27,
    23, 48, 29, 100]
 

plt.hist(data,bins=5, edgecolor = "black")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

## 2. Age Distribution with Custom Bins

```python
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## 3. Histogram with Average Line

```python
import numpy as np

average_age = np.mean(ages)

plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(average_age, color="red", linestyle='--', linewidth=2, label=f'Average: {average_age:.1f}')
plt.legend()
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
