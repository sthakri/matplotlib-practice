# Introduction to Matplotlib

This notebook covers the fundamentals of Matplotlib:
- Creating basic line plots
- Adding labels and titles
- Plotting multiple lines
- Customizing colors, markers, and line styles
- Using different plot styles
- Grid customization

## 1. Basic Line Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
```

## 2. Plot with Real Data

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

## 3. Adding Labels and Title

```python
plt.plot(years, runs)
plt.xlabel('Years')
plt.ylabel('Runs Scored')
plt.title("Sachin Tendulkar's Yearly Runs")
```

## 4. Plotting Multiple Lines with Legend

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli, label="Virat Kohli")
plt.plot(years,sehwag, label = "Virender Sehwag")

plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison')
plt.legend()
```

## 5. Custom Markers and Line Styles (Shorthand)

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

## 6. Custom Colors and Line Styles (Detailed)

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

## 7. Adding Grid

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.grid(True)
plt.legend()
```

## 10. Using the XKCD Style (Comic Style)

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.grid(True)
plt.legend()
```

```python
with plt.xkcd():
    plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
    plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
    plt.plot(years, runs, color='blue', label="Tendulkar")
    plt.grid(True)
    plt.legend()
    plt.show()
```

## 11. Example: When Plots Become Too Busy

**Note:** Keep your visualizations clean and focused. Too many overlapping lines can confuse rather than clarify.

```python
import numpy as np

for i in range(10):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
plt.show()
```
