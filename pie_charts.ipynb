# Pie Charts in Matplotlib

This notebook covers pie chart creation and customization:
- Basic pie charts
- Custom colors
- Exploded slices
- Percentage labels
- Shadow effects

## Import Libraries and Set Style

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

## 1. Basic Pie Chart

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

## 2. Pie Chart with Custom Colors

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

## 3. Pie Chart with Custom Edge Styling

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
plt.show()
```

## 4. Exploded Pie Chart (Highlighting One Slice)

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

## 5. Pie Chart with Shadow Effect

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
plt.show()
```

## 6. Rotating Start Angle

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=140
)
plt.show()
```

## 7. Pie Chart with Percentage Labels

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Career Run Share")
plt.show()
```

## 8. Example: When Pie Charts Get Crowded

**Note:** Pie charts work best with 5-7 categories. Too many slices make them hard to read.

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```
